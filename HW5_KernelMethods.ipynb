(6.15) $k\left(\mathbf{x}, {\mathbf{x}}'\right) = q\left(k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)\right)$ where $k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)$ is a valid kernel and $q\left(\mathbf{\cdot}\right)$ is a polynomial with nonnegative coefficients.<br />
<br />
Expanding the $q\left(k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)\right)$:
$$
k\left(\mathbf{x}, {\mathbf{x}}'\right) = c_{0} + c_{1}k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)+ c_{2}k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)^{2} + ... + c_{n}k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)^{n} = c_{0} + c_{1}k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)+ c_{2}k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right) + ... + c_{n}\prod^{n}k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)
$$
As each polynomial term is a product of kernels with a positive coefficient we can verify (6.15) by using (6.13), (6.17) and (6.18):<br />
<br />
(6.13) $k\left(\mathbf{x}, {\mathbf{x}}'\right) =c k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)$<br />
<br />
(6.17) $k\left(\mathbf{x}, {\mathbf{x}}'\right) = k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right) + k_{2}\left(\mathbf{x}, {\mathbf{x}}'\right)$<br />
<br />
(6.18) $k\left(\mathbf{x}, {\mathbf{x}}'\right) = k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)k_{2}\left(\mathbf{x}, {\mathbf{x}}'\right)$<br />
<br />
We can verify (6.13) by using (6.1). If $k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)$ is a valid kernel then:
$$
k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right) = \mathbf{\phi}\left(\mathbf{x}\right)^{\intercal}\mathbf{\phi}\left({\mathbf{x}}'\right)
$$
Multiplying on the left by a scalar $c$:
$$
ck_{1}\left(\mathbf{x}, {\mathbf{x}}'\right) = c\left(\mathbf{\phi}\left(\mathbf{x}\right)^{\intercal}\mathbf{\phi}\left({\mathbf{x}}'\right)\right) = \sqrt{c}\mathbf{\phi}\left(\mathbf{x}\right)^{\intercal}\sqrt{c}\mathbf{\phi}\left({\mathbf{x}}'\right)
$$
Absorbing the scalar multiplication as $\sqrt{c}\mathbf{\phi}\left(\mathbf{x}\right) = \psi$ then:
$$
ck_{1}\left(\mathbf{x}, {\mathbf{x}}'\right) = \mathbf{\psi}\left(\mathbf{x}\right)^{\intercal}\mathbf{\psi}\left({\mathbf{x}}'\right)
$$
Which by (6.1) is the definition of a valid kernel: the input vector x enters only in the form of scalar products.<br />
<br />
We can verify (6.17) by using the necessary and sufficient condition for a valid kernel is that the Gram matrix $\mathbf{K}$ must be positive semidefinite. This means that:
$\mathbf{z}^{\intercal}\mathbf{K}\mathbf{z} \geq 0$ for every nonzero real column vector $z$. Now define two Gram matrices for $k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)$ and $k_{2}\left(\mathbf{x}, {\mathbf{x}}'\right)$ as $\mathbf{K}_{1}$ and $\mathbf{K}_{2}$ respectively. Then the Gram matrix for $k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right) + k_{2}\left(\mathbf{x}, {\mathbf{x}}'\right)$ is $\mathbf{K}_{1} + \mathbf{K}_{2}$. We can then write:
$$
\mathbf{z}^{\intercal}\left(\mathbf{K}_{1} + \mathbf{K}_{2}\right)\mathbf{z} = \mathbf{z}^{\intercal}\mathbf{K}_{1}\mathbf{z} + \mathbf{z}^{\intercal}\mathbf{K}_{2}\mathbf{z} \geq 0
$$
demonstrating that $k\left(\mathbf{x}, {\mathbf{x}}'\right) = k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right) + k_{2}\left(\mathbf{x}, {\mathbf{x}}'\right)$ is a valid kernel.<br />
<br />
We can verify (6.18) by using the same process as for (6.13) above, that is, we note that iff $k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)$ and $k_{2}\left(\mathbf{x}, {\mathbf{x}}'\right)$ are a valid kernels then:
$$
k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right) = \mathbf{\phi}\left(\mathbf{x}\right)^{\intercal}\mathbf{\phi}\left({\mathbf{x}}'\right)\\
k_{2}\left(\mathbf{x}, {\mathbf{x}}'\right) = \mathbf{\psi}\left(\mathbf{x}\right)^{\intercal}\mathbf{\psi}\left({\mathbf{x}}'\right)
$$
Taking the product of the two:
$$
k\left(\mathbf{x}, {\mathbf{x}}'\right) = \mathbf{\phi}\left(\mathbf{x}\right)^{\intercal}\mathbf{\phi}\left({\mathbf{x}}'\right)\mathbf{\psi}\left(\mathbf{x}\right)^{\intercal}\mathbf{\psi}\left({\mathbf{x}}'\right)
$$
Writing the inner products as summations:
$$
k\left(\mathbf{x}, {\mathbf{x}}'\right) = \sum_{i=1}^{n_{1}}\phi_{i}\left(\mathbf{x}\right)\phi_{i}\left({\mathbf{x}}'\right)\sum_{j=1}^{n_{2}}\psi_{j}\left(\mathbf{x}\right)\psi_{j}\left({\mathbf{x}}'\right) = \sum_{i=1}^{n_{1}}\sum_{j=1}^{n_{2}}\phi_{i}\left(\mathbf{x}\right)\phi_{i}\left({\mathbf{x}}'\right)\psi_{j}\left(\mathbf{x}\right)\psi_{j}\left({\mathbf{x}}'\right)
$$
Because scalar multiplication commutes:
$$
k\left(\mathbf{x}, {\mathbf{x}}'\right) = \sum_{i=1}^{n_{1}}\sum_{j=1}^{n_{2}}\phi_{i}\left(\mathbf{x}\right)\phi_{i}\left({\mathbf{x}}'\right)\psi_{j}\left(\mathbf{x}\right)\psi_{j}\left({\mathbf{x}}'\right) = \sum_{i=1}^{n_{1}}\sum_{j=1}^{n_{2}}\phi_{i}\left(\mathbf{x}\right)\psi_{j}\left(\mathbf{x}\right)\phi_{i}\left({\mathbf{x}}'\right)\psi_{j}\left({\mathbf{x}}'\right)
$$
Create new index $m$ where $m = n_{1}\cdot n_{2}$ and combine products:
$$
k\left(\mathbf{x}, {\mathbf{x}}'\right) = \sum_{i=1}^{m}\theta_{i}\left(\mathbf{x}\right)\theta_{i}\left({\mathbf{x}}'\right) = \mathbf{\theta}\left(\mathbf{x}\right)^{\intercal}\mathbf{\theta}\left({\mathbf{x}}'\right)
$$
We have now verified (6.13), (6.17) and (6.18):<br />
<br />
(6.13) $k\left(\mathbf{x}, {\mathbf{x}}'\right) =c k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)$<br />
<br />
(6.17) $k\left(\mathbf{x}, {\mathbf{x}}'\right) = k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right) + k_{2}\left(\mathbf{x}, {\mathbf{x}}'\right)$<br />
<br />
(6.18) $k\left(\mathbf{x}, {\mathbf{x}}'\right) = k_{1}\left(\mathbf{x}, {\mathbf{x}}'\right)k_{2}\left(\mathbf{x}, {\mathbf{x}}'\right)$<br />
<br />
And therefore all components of the polynomial are valid kernels making $k\left(\mathbf{x}, {\mathbf{x}}'\right)$ of (6.15) also a valid kernel.

## Ornstein-Uhlenbeck process

An Ornstein-Uhlenbeck stochastic process can be simulated in discrete time as an autoregressive "AR(1)" model, described by the equation
$$
y(t) = (1-a)y(t-1) + b \xi(t),
$$
where $0 < a < 1$, $b > 0$, and $\xi(t) \sim \mathcal{N}(0,1)$ is drawn independently at each time step. This equation can describe, for example, the time-dependent position of a particle that gets kicked around by random forces (the second term on the right hand side) and has a restoring force toward $y=0$ (the first term on the right hand side). This model and related ones have widespread applications in the physical sciences, engineering, and financial modeling.

**Exercise 2** (3 points): Use the above equation to show that the autocorrelations are given by, for $\tau \geq 0$, $C(\tau) \equiv E[y(t) y(t-\tau)] = (1-a)^\tau C(0)$ and, for general $\tau$, $C(\tau) = (1-a)^{|\tau|} C(0)$.<br />
<br />
First note that you can iterate the Ornstein-Uhlenbeck stochastic process for $\tau$ time steps as:
$$
y(t) = (1-a)^{\tau}y\left(t - \tau \right) + \tau\cdot v(t)
$$
Where $v(t)\sim\mathcal{N}\left(0, \tau\right)$. Now note the definition of the correlation auto-correlation function between times $t$ and $t-\tau$:
$$
C(\tau) = E\left[y(t)y(t - \tau\right] =  E\left[\left\{(1-a)^{\tau}y\left(t - \tau \right) + \tau\cdot v(t)\right\}y\left(t - \tau \right)\right] = E\left[(1-a)^{\tau}y\left(t - \tau \right)y\left(t - \tau \right) + y\left(t - \tau \right) \tau\cdot v(t)\right]
$$
$$
E\left[(1-a)^{\tau}y\left(t - \tau \right)y\left(t - \tau \right)\right] + E\left[y\left(t - \tau \right) \tau\cdot v(t)\right] = E\left[(1-a)^{\tau}y\left(t - \tau \right)y\left(t - \tau \right)\right]
$$
$$
C\left(\tau\right)= (1-a)^{\tau}E\left[y\left(t - \tau \right)y\left(t - \tau \right)\right] = (1-a)^{\tau}C_{0}
$$

**Exercise 1** (5 points): Simulate an AR(1) process using the above equation for 10,000 time steps with $y(0) = 0$, $b=1$, and choosing your own value for $a$. Plot the first 100 timesteps of $y(t)$. Compute the autocorrelation $C(\tau)$ for $-100 < \tau < 100$ empirically (i.e. replacing the expectation described above by an average over $t$), and plot the empirical $C(\tau)/C(0)$ together with the theoretically derived result from above.

```python
## Solution ##
timesteps = 10000
a = 0.5
b = 1
y = np.empty(timesteps)
y[0] = 0
for i in range(1, timesteps):
    y[i] = (1 - a)*y[i-1] + b*np.random.normal(0.0, 1.0,size=(1))

fig, ax = plt.subplots()
ax.plot(y[:100])
ax.set_title('Ornstein-Uhlenbeck process simulation')
ax.set_xlabel('time (a.u.)')
ax.set_ylabel('y(t)');
plt.show()

auto_corrs = []
for tau in range(100):
    tau_corr = []
    for i in range(timesteps):
        if (i - tau > 0):
            y_corr = y[i]*y[i - tau]
            tau_corr.append(y_corr)
    auto_corrs.append(np.mean(tau_corr))

a_ = np.hstack([auto_corrs[::-1], auto_corrs])/auto_corrs[0]
fig, ax  = plt.subplots()
ax.plot(np.arange(-100, 100), a_)
ax.set_title('Ornstein-Uhlenbeck process autocorrelation')
ax.set_xlabel('tau (a.u.)')
ax.set_ylabel('Auto-correlation');
```

Rather than generating an OU process by simulating it online (i.e. one time step at a time), we can interpret it as a Gaussian process. This means that, if we have $N$ timesteps, then a curve such as the one plotted in the previous exercise is interpreted as a single sample from a $N$-dimensional multivariate Gaussian distribution: $\vec{y} \sim \mathcal{N}(\vec{y} | 0, \mathbf{K})$, where the kernel matrix has entries $K_{tt'} = \exp(-\theta|t-t'|)$, and $\vec{y} = [y(1), y(2), \ldots, y(N)]^\top$.
