In this notebook, we'll implement K-means clustering from scratch to cluster data from the Iris dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data:
from sklearn.datasets import load_iris
data = load_iris()  # data is a dictionary containing lots of things

# List of the features and targets being measured:
print('Features: \n', data['feature_names'])
print('Targets: \n', data['target_names'])

# The input data (each row gives measurements of the features listed above):
X = data['data']
n_samples, n_features = np.shape(data['data'])
print('n_samples, n_features: ', (n_samples, n_features))

# The target data (subspecies identity):
y = data['target']

# Print the first few samples of x and y data:
print('X data: \n', X[:3,:])
print('y data: \n', y[:3])
```

The main output that we care about is `cluster_IDs`, a 1D array of integers (0, 1, ..., K-1) denoting the cluster assignment of each data point.

```python
# Number of clusters:
K = 3  

# Initial guesses of cluster means are randomly chosen data points:
mu = X[np.random.choice(n_samples, K), :]  # mu is an array of shape K-by-n_features

# The K-means algorithm:
n_iter = 10
cluster_IDs = np.zeros(n_samples)
losses = []
for i in range(n_iter):
    # Assign each data point to the nearest mean vector (E step):
    ### YOUR CODE HERE ###
    for n in range(n_samples):
        distances = [np.linalg.norm(X[n,:]-mu[k,:]) for k in range(K)]
        cluster_IDs[n] = np.argmin(distances)

    # Recompute the mean vectors (M step):
    ### YOUR CODE HERE ###
    for k in range(K):
        mask = cluster_IDs == k
        mu[k,:] = np.mean(X[mask,:],axis=0)#take the mean of each cluster


    # Compute the loss function:
    ### YOUR CODE HERE ###
    J = 0
    for k in range(K):
        J += np.linalg.norm(X[cluster_IDs==k,:] - mu[k,:])**2
    losses.append(J)

# Plot the loss function:
plt.plot(losses, '-o')
plt.xlabel('Iteration')
plt.ylabel('Loss')
```

To do this in a way that's more generally useful, we can wrap the above block of code into a function that implements the algorithm.

**Exercise**: Turn the above block of code into a function, `k_means(X, K)`, which returns `cluster_IDs`.

```python
### YOUR CODE HERE ###
class k_means():
    def __init__(self, X, K, n_iter):
        self.K = K
        self.X = X
        self.n_iter = n_iter
    
    def fit(self, plot=True):
        '''
        Implement the K-means clustering algorithm on data X
        '''
        
        # Initial guesses of cluster means are randomly chosen data points:
        mu = self.X[np.random.choice(n_samples, self.K), :]  # mu is an array of shape K-by-n_features

        # The K-means algorithm:
        cluster_IDs = np.zeros(n_samples)
        losses = []
        for i in range(self.n_iter):
            # Assign each data point to the nearest mean vector (E step):
            ### YOUR CODE HERE ###
            for n in range(n_samples):
                distances = [np.linalg.norm(self.X[n,:]-mu[k,:]) for k in range(self.K)]
                cluster_IDs[n] = np.argmin(distances)

            # Recompute the mean vectors (M step):
            ### YOUR CODE HERE ###
            for k in range(self.K):
                mask = cluster_IDs == k
                mu[k,:] = np.mean(self.X[mask,:],axis=0)#take the mean of each cluster


            # Compute the loss function:
            ### YOUR CODE HERE ###
            J = 0
            for k in range(K):
                J += np.linalg.norm(self.X[cluster_IDs==k,:] - mu[k,:])**2
            losses.append(J)

        if plot:
            # Plot the loss function:
            plt.plot(losses, '-o')
            plt.xlabel('Iteration')
            plt.ylabel('Loss')
                
        self.labels_ = cluster_IDs

kmeans_model = k_means(X, K, n_iter=10)
kmeans_model.fit()
kmeans_model.labels_
```
